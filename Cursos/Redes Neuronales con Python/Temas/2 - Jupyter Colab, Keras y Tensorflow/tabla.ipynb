Entrenamos esta red neuronal sencilla para que encuentre la tabla de multiplicación que vincula los datos de entrada y las etiquetas.


```python
# Importamos a nuestra libreta de trabajo las librerias a utilizar
import tensorflow as tf # libreria para IA desarrollada por google
import numpy as np # para trabajar con números
import matplotlib.pyplot as plt # se importa la libreria matplot para hacer las gráficas
```

# Definimos el Corpus

```python
# Genera datos de entrenamiento
numeros = np.random.randint(1, 11, size=(1000,))  # Genera 1000 números aleatorios del 1 al 10
resultados = numeros * 9  # Multiplica cada número por 7 para obtener los resultados deseados
```

# Creamos una Capa de Percetrones

```python
# Crea el modelo de red neuronal

model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])

# Se crea una capa densa en una red neuronal con una sola neurona de salida.
# Por lo general, esta configuración se utiliza en problemas de regresión, donde
# se trata de predecir un valor continuo a partir de una sola característica de entrada.


```

# Compilamos el modelo y generamos el resumen

```python
#  COMPILAMOS el MODELO. Se prepara el modelo para ser entrenado. Se utilizan dos propiedades: optimizador y función de pérdida.
#  Optimizador utilizamos el algoritmo de Adams, permite a la red saber como ajustar PESOS y SESGOS para ir mejorando

model.compile(optimizer='adam', loss='mean_squared_error')

#   Aquí se especifican dos aspectos clave del proceso de entrenamiento de la red neuronal:

#   Optimizador (optimizer): Se ha utilizado el optimizador "Adam". Responsable de ajustar los pesos de la red neuronal durante el entrenamiento
#   para minimizar la función de pérdida. "Adam" es un optimizador popular y eficaz utilizado en una variedad de tareas de aprendizaje profundo.#

#   Función de Pérdida (loss): La función de pérdida se establece en "mean_squared_error" (error cuadrático medio).
#   Esta función de pérdida se utiliza comúnmente en problemas de regresión, donde se trata de predecir un valor numérico continuo.
#   El objetivo del modelo durante el entrenamiento es minimizar esta función de pérdida, lo que implica que el modelo está tratando de
#   hacer que sus predicciones se ajusten lo más cerca posible a los valores reales.

#   Una vez compilado el modelo de esta manera, está listo para ser entrenado utilizando datos de entrenamiento mediante el método fit.
#   El optimizador y la función de pérdida se utilizarán durante el proceso de entrenamiento para ajustar los pesos de la red neuronal
#   y hacer que haga predicciones más precisas en función de los datos de entrenamiento.


```

```python
# Muestra un resumen de la arquitectura del modelo
model.summary()
```

# Entrenamos la *neurona*

```python

# Entrena el modelo
history = model.fit(numeros, resultados, epochs=1000)

# Evalúa el modelo entrenado
loss = model.evaluate(numeros, resultados)
print(f"Pérdida final del modelo: {loss}")

# Hacer predicciones con nuevos datos
nuevo_numero = np.array([3, 5, 9, 7, 11])  # Por ejemplo, prueba con estos números
predicciones = model.predict(nuevo_numero)
print("Predicciones:")
for i in range(len(nuevo_numero)):
    print(f"Para {nuevo_numero[i]}, la red predice {predicciones[i][0]:.2f}")
```

# Graficamos la función de pérdida

```python
# RESULTADO DE LA FUNCIÓN DE PÉRDIDA. Nos informa como de mal están los resultados en cada vuelta
import matplotlib.pyplot as plt # se importa la libreria matplot para hacer las gráficas
plt.xlabel("# Vuelta dada")
plt.ylabel("Magnitud de pérdida")
plt.plot(history.history["loss"])
```

```python
import numpy as np

# Definir nuevos números para hacer predicciones
nuevos_numeros = np.array([3, 5, 9, 7, 11])  # Debe ser una matriz 2D

# Realizar predicciones con el modelo
predicciones = model.predict(nuevos_numeros)

# Calcular la tabla de multiplicar en función de la división entre la predicción y el primer número
def determinar_tabla_de_multiplicar(prediccion, numero):
    if numero == 0:
        return "No se puede dividir por cero"

    tabla = prediccion / numero
    return round(tabla)

# Imprimir las predicciones y la tabla de multiplicar correspondiente
print("Predicciones:")
for i in range(len(nuevos_numeros)):
    prediccion = predicciones[i][0]
    primer_numero = nuevos_numeros[i][0]

    tabla_de_multiplicar = determinar_tabla_de_multiplicar(prediccion, primer_numero)

    if isinstance(tabla_de_multiplicar, int):
        print(f"Para {nuevos_numeros[i]}, la red predice {prediccion:.2f}, que corresponde a la tabla de multiplicar {tabla_de_multiplicar}.")
    else:
        print(f"Para {nuevos_numeros[i]}, la red predice {prediccion:.2f}, {tabla_de_multiplicar}.")
```
